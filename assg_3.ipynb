```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Basics of Probability

### a) Coin Toss Simulation
We simulate tossing a fair coin 10,000 times and compute experimental probabilities.

```python
trials = 10000
heads = 0
tails = 0

for _ in range(trials):
    if random.choice(['H', 'T']) == 'H':
        heads += 1
    else:
        tails += 1

print('Probability of Heads:', heads / trials)
print('Probability of Tails:', tails / trials)
```

### b) Rolling Two Dice
We roll two dice repeatedly and estimate the probability of getting a sum of 7.

```python
trials = 10000
count_sum7 = 0

for _ in range(trials):
    d1 = random.randint(1, 6)
    d2 = random.randint(1, 6)
    if d1 + d2 == 7:
        count_sum7 += 1

print('Probability of sum = 7:', count_sum7 / trials)
```

## 2. Probability of At Least One 6 in 10 Rolls

We estimate the probability using repeated simulations.

```python
def prob_at_least_one_six(trials=10000):
    success = 0
    for _ in range(trials):
        rolls = [random.randint(1, 6) for _ in range(10)]
        if 6 in rolls:
            success += 1
    return success / trials

print('Estimated Probability:', prob_at_least_one_six())
```

## 3. Conditional Probability and Bayesâ€™ Theorem

A bag contains red, green, and blue balls. We simulate repeated draws with replacement.

```python
balls = ['R'] * 5 + ['G'] * 7 + ['B'] * 8
trials = 1000

previous = None
blue_then_red = 0
previous_blue = 0

for _ in range(trials):
    current = random.choice(balls)
    if previous == 'B':
        previous_blue += 1
        if current == 'R':
            blue_then_red += 1
    previous = current

print('P(Red | Previous Blue):', blue_then_red / previous_blue)

# Bayes Theorem Verification
p_red = 5 / 20
p_blue = 8 / 20
p_blue_given_red = p_blue
p_red_given_blue = (p_blue_given_red * p_red) / p_blue

print('Bayes Theorem Result:', p_red_given_blue)
```

## 4. Discrete Random Variable

We generate samples from a given discrete distribution and compute statistics.

```python
values = [1, 2, 3]
probabilities = [0.25, 0.35, 0.4]

sample = np.random.choice(values, size=1000, p=probabilities)

print('Mean:', np.mean(sample))
print('Variance:', np.var(sample))
print('Standard Deviation:', np.std(sample))
```

## 5. Exponential Distribution

We simulate data and visualize it using a histogram and PDF overlay.

```python
data = np.random.exponential(scale=5, size=2000)

plt.hist(data, bins=30, density=True)

x = np.linspace(0, max(data), 200)
pdf = (1/5) * np.exp(-x/5)
plt.plot(x, pdf)

plt.title('Exponential Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

## 6. Central Limit Theorem

We compare a uniform distribution with the distribution of sample means.

```python
uniform_data = np.random.uniform(0, 1, 10000)

sample_means = []
for _ in range(1000):
    sample = np.random.choice(uniform_data, size=30)
    sample_means.append(np.mean(sample))

plt.hist(uniform_data, bins=30)
plt.title('Uniform Distribution')
plt.show()

plt.hist(sample_means, bins=30)
plt.title('Distribution of Sample Means')
plt.show()
```
