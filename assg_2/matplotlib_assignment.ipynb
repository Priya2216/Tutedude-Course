This notebook demonstrates basic data visualization techniques.

```python
import matplotlib.pyplot as plt
import numpy as np
```

## 1. Line Plot

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]

plt.plot(x, y)
plt.title('Line Plot Example')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.grid(True)
plt.show()
```

## 2. Bar Graph

```python
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]

plt.bar(students, marks)
plt.title('Student Marks')
plt.xlabel('Students')
plt.ylabel('Marks')
plt.show()
```

## 3. Pie Chart

```python
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]

explode = (0.1, 0, 0, 0)
plt.pie(revenue, labels=regions, explode=explode, autopct='%1.1f%%')
plt.title('Revenue Distribution')
plt.show()
```

## 4. Histogram

```python
data = np.random.randint(1, 101, 1000)
plt.hist(data)
plt.title('Histogram of Random Numbers')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
