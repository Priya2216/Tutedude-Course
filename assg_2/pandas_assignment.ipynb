This notebook covers DataFrame creation, filtering, grouping, and basic data analysis.

```python
import pandas as pd
```

## 1. Create DataFrame and Basic Analysis

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [24, 27, 22, 32, 29],
    'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'],
    'Salary': [45000, 54000, 50000, 62000, 47000]
}

df = pd.DataFrame(data)
df
```

```python
print('Summary Statistics:')
print(df[['Age', 'Salary']].describe())

avg_hr_salary = df[df['Department'] == 'HR']['Salary'].mean()
print('Average HR Salary:', avg_hr_salary)
```

## 2. Add Bonus Column

```python
df['Bonus'] = df['Salary'] * 0.10
df
```

## 3. Filter Employees Aged Between 25 and 30

```python
filtered_df = df[(df['Age'] >= 25) & (df['Age'] <= 30)]
filtered_df
```

## 4. Group by Department

```python
grouped = df.groupby('Department')['Salary'].mean()
grouped
```

## 5. Sort by Salary and Save to CSV

```python
sorted_df = df.sort_values(by='Salary')
sorted_df.to_csv('sorted_salary.csv', index=False)
sorted_df
```
