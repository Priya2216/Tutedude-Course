# Assignment 1 – Working with NumPy

This notebook demonstrates basic NumPy array creation, manipulation, and mathematical operations.

```python
import numpy as np
```

## 1. 1D Array Operations
Create a 1D array from 1 to 20 and compute basic statistics.

```python
arr = np.arange(1, 21)
print('Array:', arr)

print('Sum:', np.sum(arr))
print('Mean:', np.mean(arr))
print('Median:', np.median(arr))
print('Standard Deviation:', np.std(arr))

indices = np.where(arr > 10)
print('Indices of elements > 10:', indices)
```

## 2. 2D Array Operations
Create a 4×4 array and perform transpose and sum operations.

```python
arr2d = np.arange(1, 17).reshape(4, 4)
print('2D Array:\n', arr2d)

print('Transpose:\n', arr2d.T)

print('Row-wise sum:', arr2d.sum(axis=1))
print('Column-wise sum:', arr2d.sum(axis=0))
```

## 3. Operations on Two 3×3 Random Arrays

```python
a = np.random.randint(1, 21, (3, 3))
b = np.random.randint(1, 21, (3, 3))

print('Array A:\n', a)
print('Array B:\n', b)

print('Addition:\n', a + b)
print('Subtraction:\n', a - b)
print('Multiplication:\n', a * b)

print('Dot Product:\n', np.dot(a, b))
```

## 4. Reshaping and Slicing

```python
arr1d = np.arange(1, 13)
reshaped = arr1d.reshape(3, 4)
print('Reshaped Array:\n', reshaped)

sliced = reshaped[:2, -2:]
print('Sliced Array:\n', sliced)
```
